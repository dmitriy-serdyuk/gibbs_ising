```python
import main
import matplotlib
%matplotlib inline
from matplotlib import pyplot as plt
import numpy as np
import matplotlib.cm as cm
```

```python
num_iter = 1000
size = 30
theta = 0.45
show_freq = 100
ans = main.main(num_iter, size, theta, show_freq)
```

```python
plt.figure(1, figsize=(15, 6))
for i, a in enumerate(ans):
    plt.subplot(250 + i)
    plt.imshow(a, interpolation='none', cmap=cm.Greys_r)
plt.savefig("samples%d.eps" % (int(theta * 100)))
plt.show()
```
