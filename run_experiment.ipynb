```python
import main
import block_gibbs
import matplotlib
%matplotlib inline
from matplotlib import pyplot as plt
import numpy as np
import matplotlib.cm as cm
import cPickle as pkl
```

```python
num_iter = 1000
size = 60
theta = 0.25
show_freq = 100
ans = block_gibbs.sample(num_iter=num_iter, size=size, theta=theta, vis_step=show_freq)
```

```python
plt.figure(1, figsize=(15, 6))
for i, a in enumerate(ans):
    plt.subplot(250 + i)
    plt.imshow(a, interpolation='none', cmap=cm.Greys_r)
plt.savefig("samples%d.eps" % (int(theta * 100)))
plt.show()
```

```python
with open('samples.pkl', 'w') as f:
    pkl.dump(ans, f)
```

```python
with open('samples.pkl', 'r') as f:
    samples = pkl.load(f)
```

```python
thetas = block_gibbs.infer(data=samples, start_theta=0.99, grad_step=0.00001)
```

```python
from itertools import tee, izip
def pairwise(iterable):
    "s -> (s0,s1), (s1,s2), (s2, s3), ..."
    a, b = tee(iterable)
    next(b, None)
    return izip(a, b)

for theta_prev, theta in pairwise(thetas):
    print theta
    if mod(theta_prev - theta) < 0.0001:
        break
```
