```python
import main
import block_gibbs
import matplotlib
%matplotlib inline
from matplotlib import pyplot as plt
import numpy as np
import matplotlib.cm as cm
import cPickle as pkl
```

```python
num_iter = 10000
size = 30
theta = 0.25
show_freq = 100
ans = block_gibbs.sample(num_iter=num_iter, size=size, theta=theta, vis_step=show_freq)
```

```python
plt.figure(1, figsize=(15, 6))
for i, a in enumerate(ans):
    plt.subplot(250 + i)
    plt.imshow(a, interpolation='none', cmap=cm.Greys_r)
plt.savefig("samples%d.eps" % (int(theta * 100)))
plt.show()
```

```python
with open('samples.pkl', 'w') as f:
    pkl.dump(ans, f)
```

```python
with open('samples.pkl', 'r') as f:
    data = pkl.load(f)
len(data)
```

```python
thetas = block_gibbs.infer(data=data, start_theta=1.0, grad_step=0.0003, sample_iter=100, sample_freq=10)
```

```python
from itertools import tee, izip
def pairwise(iterable):
    "s -> (s0,s1), (s1,s2), (s2, s3), ..."
    a, b = tee(iterable)
    next(b, None)
    return izip(a, b)
```

```python
ans = []
for theta_prev, theta in pairwise(thetas):
    print theta
    ans +=  [theta]
    if abs(theta_prev - theta) < 1e-5:
        break
```

```python
ans = np.array(ans)
line, = plt.plot((ans - 0.25) ** 2)
plt.xlabel('Epoch')
plt.ylabel('Square error')
plt.legend([line], ['gradiend step 0.0003,\ndecrease by 3% every epoch'])
plt.yscale('log')
plt.savefig('all_dataset.png')
```

```python
with open('results.csv', 'w') as f:
    print >> f, ans
```
